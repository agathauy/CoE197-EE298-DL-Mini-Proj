```python

import pretty_midi
import matplotlib.pyplot as plt
import os
import lib.file_util
from lib.midi_util import midi_to_array_one_hot, stylify
from mido import MidiFile
from random import shuffle
```

```python
mid_path = './data/preprocessed/single_piano/midi_bts'
```

```python
for i , filename in enumerate(os.listdir(mid_path)):
    print(filename)
    
    try:
        midi_data = pretty_midi.PrettyMIDI(os.path.join(mid_path, filename))
        mid = MidiFile(os.path.join(mid_path, filename))
    except (KeyError, IOError, IndexError, EOFError, ValueError):
        print("NAUGHTY")
        break
        
    midi_array, velocity_array = midi_to_array_one_hot(mid, 4)
    
    print("[MIDI_ARRAY]")
    print(midi_array)
    print(midi_array.shape)
    
    print("[VELOCITY_ARRAY]")
    print(velocity_array)
    print(velocity_array.shape)
    # Just test one file
    break
```


# Attempt to convert back to midi 
Assuming midi_array is the prediction of the LSTM

```python
revert_mid = stylify(mid, velocity_array/127, 4) # divde by 127 since this accepts a mapped output
revert_mid.save(os.path.join("output", "revert", filename))
```
